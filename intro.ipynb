### 3. Calculate the quantities using mathematical formula and compare with the original data to ensure understanding 计算并对比重要的数值（如死亡率等）

Formula Sheet

$L(x)= [l(x) - \frac{d(x)}{2}$] * 1    person-year

$T(x)= \sum_{y=x}^\infty L(x) = L(x) + T(x+1)$

$e(x) = \frac{T(x)}{l(x)}$

```python
orig_pop = 100000   #l_0
```

```python
# Calculate L(x): person-years lived in the interval
GBR1990['my_L(x)'] = GBR1990['l(x)'] - GBR1990['d(x)'] / 2

# Calculate T(x): total person-years above age x
GBR1990['my_T(x)'] = GBR1990['my_L(x)'][::-1].cumsum()[::-1]  # Reverse, cumulative sum, then reverse back

# Calculate e(x): life expectancy at age x
GBR1990['my_e(x)'] = GBR1990['my_T(x)'] / GBR1990['l(x)']
```
